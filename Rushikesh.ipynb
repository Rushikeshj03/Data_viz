# **Univariate**


**1.What is market share for top 10 Platforms**

```python
# @title
gradable_counts=df.Platform.value_counts().nlargest(10)
# Plot a pie chart
gradable_counts.plot(kind='pie', title='Top 10 Platforms', figsize=(8,8),autopct='%.2f %%')
plt.legend()
plt.show()
```



Here I have created a pie chart of the percentage of the top 10 selected platforms. Note the platforms DS, PS2, PS3, and Wii have the largest percentage of games, and platforms PC, XB, and GBA have the lowest percentage of games



**2.Which is most selled video game genre**

```python
# @title
#creating a pie chart
genre = df.Genre.value_counts().values
labels =df.Genre.value_counts().index

plt.title('Genre by Sales',fontdict={'fontweight':'bold'})

plt.pie(genre,labels=labels,autopct='%.2f %%',pctdistance=0.9)

plt.legend(bbox_to_anchor=(1.5,1.0),loc='upper left')

plt.show()
```


Insights:

The 20.44% of the video games that is selled belongs to the Action genre,followed by sports(13.33%).The least selled video games belong to the puzzle Genre **bold text**

**3.Which years has the top 10 sales on video games.**

```python
# @title
#creating a bar chart
year= df.Year.value_counts().head(10)
plt.figure(figsize=(12,8))
print(year)
plt.xticks(year.index)
plt.title('Top 10 sales by Years',fontdict={'fontweight':'bold','fontsize':16})

plt.bar(year.index,year.values)
plt.show()
```

Conclusion:

The highest count is for the year 2007, indicating that this year has the highest occurrence in the dataset. The counts then generally decrease for subsequent years, with 2014 having the same count as 2006.

Understanding the distribution of data across different years is important for temporal analysis and identifying patterns or trends over time. Businesses can use this information to assess performance variations across different years and make informed decisions based on historical data.



**4.Which platform has the highest sales among all.**

```python
# @title
#creating a horizontal  bar  chart
platform = df.Platform.value_counts()
plt.figure(figsize=(12,8))

plt.barh(platform.index,platform.values)

for i,v in enumerate(platform.values):
    plt.text(v,i,str(v))

plt.title('Sales by Platform',fontdict={'fontweight':'bold','fontsize':'14'})
plt.xlabel('Counts')
plt.ylabel("Platform")

plt.show()
```


Conclusion:
The list provides a distribution of video game platforms and their respective counts. It seems to represent the popularity or prevalence of different gaming platforms. Platforms like PS2, X360, and PS3 have higher counts, indicating their prominence in the dataset, while others have lower counts.

Understanding the popularity of platforms is crucial for game developers, publishers, and retailers, as it helps them make informed decisions about game development, marketing, and sales strategies tailored to specific platforms.


**5.Plot the sales using a horizontal bar graph**

```python
# @title

a = df.NA_Sales.sum().round()
b = df.EU_Sales.sum().round()
c = df.JP_Sales.sum().round()
d = df.Other_Sales.sum().round()
e = df.Global_Sales.sum().round()

values = [a,b,c,d,e]
labels=['North America','Europe','Japan','Other Sales','Global Sales']

plt.figure(figsize=(12,8))

plt.xlabel("Sales By Countries")
plt.ylabel('Sales(in millions)')
plt.bar(labels,values)

for i,v in enumerate(values):
    plt.text(i,v+1,str(v),ha='center')

plt.show()
```


Insights:

North America Dominates the sales by country wise.It contributes half of the global sales


```python
# @title
# Setting up a new figure with specified dimensions
plt.figure(figsize=(15, 5))

# Creating a count plot to visualize the number of games released each year
# The 'order' parameter ensures the years are displayed in chronological order
sns.countplot(data=df, x='Year', order=df['Year'].value_counts().index.sort_values())

# Setting the title and labels for the plot for better clarity
plt.title('Number of Games Released by Year')
plt.xlabel('Year of Release')
plt.ylabel('Number of Games Released')

# Rotating the x-axis labels for better readability
plt.xticks(rotation=45)

# Adjusting the layout to ensure all elements are displayed correctly
plt.tight_layout()

# Displaying the plot
plt.show()
```

 **Top 10 publishers**

```python
# @title
# Group the data by "Publisher" and sum the "Global_Sales" column
publisher_sales = df.groupby('Publisher')['Global_Sales'].sum().reset_index()

# Sort the data by global sales in descending order to find top publishers
top_publishers = publisher_sales.sort_values(by='Global_Sales', ascending=False).head(10)

# Create a bar chart
plt.figure(figsize=(12, 6))
plt.bar(top_publishers['Publisher'], top_publishers['Global_Sales'], color='skyblue')
plt.xlabel('Publisher')
plt.ylabel('Global Sales (in millions)')
plt.title('Top 10 Publishers by Global Sales')
plt.xticks(rotation=90)
plt.tight_layout()

# Show the chart
plt.show()
```

# **Bivariate**

```python
# @title
# Creating a scatter plot to visualize the relationship between NA Sales and Global Sales
df.plot(kind='scatter', x='NA_Sales', y='Global_Sales', alpha=0.5, color='red', figsize=(5,5))

# Setting labels and title for the plot for better clarity
plt.xlabel('NA-Sales')  # x-axis label
plt.ylabel('Global-Sales')  # y-axis label
plt.title('Linear relationship graph analysis between NA-Sales and Global-Sales')  # plot title

# Adding an annotation to the scatter plot
max_na_sales = df['NA_Sales'].max()
max_global_sales = df['Global_Sales'].max()
plt.annotate('Strong 1:1 correlation observed',
             xy=(max_na_sales * 0.5, max_global_sales * 0.5),
             xytext=(max_na_sales * 0.3, max_global_sales * 0.7),
             arrowprops=dict(facecolor='black', arrowstyle='->'),
             fontsize=9)

# Display the modified plot
plt.show()
```

```python
# @title


# Select the variables for the scatter plot
x = df['NA_Sales']
y = df['EU_Sales']
global_sales = df['Global_Sales']

# Create a scatter plot with a colorbar
plt.figure(figsize=(10, 6))
sc = plt.scatter(x, y, c=global_sales, cmap='viridis', alpha=0.5)
plt.xlabel('NA_Sales')
plt.ylabel('EU_Sales')
plt.title('Scatter Plot: NA_Sales vs. EU_Sales with Colorbar (Global_Sales)')
plt.colorbar(sc, label='Global_Sales')
plt.show()
```

```python
# @title
# Stacked Bar Chart
genre_platform_sales = df.pivot_table(index='Platform', columns='Genre', values='Global_Sales', aggfunc='sum')

genre_platform_sales.plot(kind='bar', stacked=True, figsize=(12, 6))
plt.title('Stacked Bar Chart: Sales by Genre and Platform')
plt.xlabel('Platform')
plt.ylabel('Total Global Sales')
plt.xticks(rotation=90)
plt.show()
```

Conclusion:
The table presents the total sales (in million units) for different video game genres on various gaming platforms. Each cell represents the total sales of a specific genre on a specific platform. This information is valuable for understanding the popularity of different genres on different platforms, which can guide game developers, publishers, and marketers in tailoring their strategies to maximize sales in specific market segments.

```python
# @title
publisher_sales = df.groupby('Publisher')['Global_Sales'].sum().reset_index()
plt.figure(figsize=(12, 6))
sns.scatterplot(x='Publisher', y='Global_Sales', size='Global_Sales', data=publisher_sales)
plt.title('Bubble Chart: Global Sales by Publisher')
plt.xticks(rotation=90)
plt.xlabel('Publisher')
plt.ylabel('Total Global Sales')
plt.show()
```

Conclusion:
The data reveals the global sales of various video game publishers, with Nintendo leading the chart with significant sales, followed by Electronic Arts, Activision, and others. This information is crucial for stakeholders in the gaming industry to understand the market dynamics and make informed decisions about partnerships, investments, and game development strategies.

# **Multivariate**

```python
# @title
yearly_sales = df.groupby('Year')[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].sum()

# Create a line plot
yearly_sales.plot(kind='line', marker='o')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.title('Yearly Sales by Region')
plt.legend(loc='upper left')
plt.show()
```

The provided data presents the video game sales (in million units) across different regions (North America, Europe, Japan, and Other regions) for each year from 1980 to 2016. Here are some notable trends and observations:

Peak Years:

The data indicates a surge in video game sales starting from the mid-1990s, with peak years observed in the late 2000s.
The years 2006 and 2007 stand out as the peak years, with substantial sales in all regions.
Regional Disparities:

North America consistently leads in video game sales, followed by Europe and Japan.
The "Other" category represents sales in regions outside North America, Europe, and Japan.
1990s Boom:

The 1990s witnessed a significant increase in global video game sales, with each region experiencing growth.
Global Decline:

After the peak years in the late 2000s, there is a noticeable decline in global sales. This could be attributed to factors such as changes in gaming platforms, market saturation, or economic conditions.
Recent Trends:

In more recent years (2014 to 2016), there is a visible decline in sales, possibly indicating a shift in the gaming industry landscape.

**Correlations between features in dataset**

```python
# @title
#Recall import seaborn as sns for use seaborn
corr_matrix = df.corr()
top_corr_features = corr_matrix.index

plt.figure(figsize=(8,6))
#plot heat map
g=sns.heatmap(df[top_corr_features].corr(),annot=True,cmap="RdYlGn")
plt.title("Correlations between features in dataset")
```

Global Sales Correlations:

The "Global_Sales" column, representing the total global sales, has strong positive correlations with:
NA_Sales (North American sales)
EU_Sales (European sales)
Other_Sales (sales in other regions)
These correlations suggest that as sales increase in one region, they are likely to increase globally.
Regional Sales Correlations:

NA_Sales and EU_Sales have a strong positive correlation, indicating that sales in North America and Europe tend to move together.
JP_Sales (sales in Japan) show a moderate positive correlation with both NA_Sales and EU_Sales.
Year and Global Sales:

There is a very weak negative correlation between the "Year" and "Global_Sales." This suggests that, overall, there is no clear linear trend between the year of release and global sales.
Rank and Global Sales:

The "Rank" column, representing the rank of games, has a negative correlation with global sales. This means that games with lower ranks (higher popularity) tend to have higher global sales.
In summary, the correlation matrix highlights the strong influence of regional sales on global sales. Additionally, it suggests that the popularity of games, as indicated by lower ranks, is associated with higher global sales. However, the year of release does not show a strong linear correlation with global sales.
